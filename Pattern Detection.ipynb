```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_flat = X_train.reshape(len(X_train),28 * 28)
```

```python
X_train_flat.shape
```

```python
X_test_flat = X_test.reshape(len(X_test),28 * 28)
```

```python
X_test_flat.shape
```

```python
X_train_flat[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']

)

model.fit(X_train_flat, y_train, epochs = 5)
```

```python
model.evaluate(X_test_flat, y_test)
```

```python
plt.matshow(X_test[1])
```

```python
y_predicted = model.predict(X_test_flat)
y_predicted[0]
```

```python
np.argmax(y_predicted[1])
```

```python
y_labels_predicted = [np.argmax(i) for i in y_predicted]
y_labels_predicted[:5]
```

```python
y_test[:5]
```

```python
cm = tf.math.confusion_matrix(labels = y_test,predictions = y_labels_predicted )
cm
```

```python
import seaborn as sns
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
# WITH HIDDEN LAYER

model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation = 'relu'), # Starting should be less than input
    keras.layers.Dense(10,activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']

)

model.fit(X_train_flat, y_train, epochs = 5)
```

```python
model.evaluate(X_test_flat, y_test)
```

```python
y_predicted = model.predict(X_test_flat)
y_labels_predicted = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels = y_test,predictions = y_labels_predicted )

import seaborn as sns
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
# ADDING FLATTEN FROM KERAS

model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, activation = 'relu'), # Starting should be less than input
    keras.layers.Dense(10,activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']

)

model.fit(X_train, y_train, epochs = 5)
```
